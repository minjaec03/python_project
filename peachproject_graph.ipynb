```python
def compare_sj_graph(title,x,x1,y):   #세종시와 비교하는 그래프 그리는 함수 
    plt.tight_layout()
    plt.subplot(4, 2, i+1) 
    plt.title(title[1] +':'+ title[i+2])
    plt.plot(x, y[1],
             color='skyblue',
             marker='o', markerfacecolor='blue',
             markersize=6)
    plt.ylabel(title[1], color='blue')
    plt.tick_params(axis='y', labelcolor='blue')


    
    y_right = plt.twinx()

    if i == 5:
       
        y_right.plot(x1, y[i+2],
                 color='pink',
                 marker='o', markerfacecolor='red',
                 markersize=6)
    else: 
        y_right.plot(x, y[i+2],
                 color='pink',
                 marker='o', markerfacecolor='red',
                 markersize=6)
            
    plt.ylabel(title[i+2], color='red')
    y_right.tick_params(axis='y', labelcolor='red')
```

```python
def corr_line_bar_graph(title1, title2, x, x1, y1, y2):  #직선과 직선의 그래프 그리기 
    plt.tight_layout()
    plt.subplot(5, 2, i+1) 
    plt.title(spot1[i])
    
    if i == 7:
        plt.plot(x1, y1[i],
                 color='red',
                 marker='o', markerfacecolor='red',
                 markersize=6)
    else : 
        plt.plot(x, y1[i],
                 color='red',
                 marker='o', markerfacecolor='red',
                 markersize=6) 
    plt.ylabel(title1, color='red')
    plt.tick_params(axis='y', labelcolor='red')

    
    y_right = plt.twinx()
     
    if i == 7:
        y_right.bar(x1,y2[i],color = 'b', width = .5)
    else: 
        y_right.bar(x,y2[i],color = 'b', width = .5)
        
   
    plt.ylabel(title2, color='blue')
    y_right.tick_params(axis='y', labelcolor='blue')
```

```python
def corr_line_line_graph(title1, title2, x, x1, y1, y2):  #직선과 바 그래프 그리기 
    plt.tight_layout()
    plt.subplot(5, 2, i+1) 
    plt.title(spot1[i])
    
    if i == 8:
        plt.plot(x1, y1[i],
                 color='red',
                 marker='o', markerfacecolor='red',
                 markersize=6)
    else : 
        plt.plot(x, y1[i],
                 color='red',
                 marker='o', markerfacecolor='red',
                 markersize=6) 
    plt.ylabel(title1, color='red')
    plt.tick_params(axis='y', labelcolor='red')

    
    y_right = plt.twinx()
     
    if i == 8:
        y_right.plot(x1, y2[i],
             color='blue',
             marker='o', markerfacecolor='blue',
             markersize=6)

    else: 
        y_right.plot(x, y2[i],
             color='blue',
             marker='o', markerfacecolor='blue',
             markersize=6)
   
   
    plt.ylabel(title2, color='blue')
    y_right.tick_params(axis='y', labelcolor='blue')

    
    
```
